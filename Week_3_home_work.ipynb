```python
#Grading
df.assign(Percent = lambda x: (x['Total'] / 300) * 100).round()
```

```python
def grade(score):
    if score >= 90:
        return 'A'
    elif score >= 70:
        return 'B'

    elif score >= 50:
        return 'C'

    else:
        return 'F'

```

```python
df['Grade']= df['Total'].apply(grade)
df
```
